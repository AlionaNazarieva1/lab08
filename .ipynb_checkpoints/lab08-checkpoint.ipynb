```python
plt.plot(t,s[:,0],'r', linewidth=2.0,label="первая фирма")
plt.plot(t,s[:,1],'b', linewidth=2.0,label="вторая фирма")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import math
from scipy.integrate import odeint
import matplotlib.pyplot as plt
p_cr = 45
tau1 = 21
p1 = 8
tau2 = 18
p2 = 13
V = 30
q = 1
a1 = p_cr/(tau1*tau1*p1*p1*V*q)
a2 = p_cr/(tau2*tau2*p2*p2*V*q)
b = p_cr/(tau1*tau1*tau2*tau2*p1*p1*p2*p2*V*q)
c1 = (p_cr-p1)/(tau1*p1)
c2 = (p_cr-p2)/(tau2*p2)
t0 = 0
tmax = 30
dt = 0.01
def dy(s,t):
    dy1 =  (c1/c1)*s[0] - (a1/c1)*s[0]*s[0] - ((b/c1)+0.00026)*s[0]*s[1]
    dy2 =  (c2/c1)*s[1] - (a2/c1)*s[1]*s[1] - (b/c1)*s[0]*s[1]
    return [dy1, dy2]
t = np.arange(t0,tmax,dt) 
v0=[2,1]
s = odeint(dy,v0,t)
```

```python
plt.plot(t,s[:,0],'r', linewidth=2.0,label="первая фирма")
plt.plot(t,s[:,1],'b', linewidth=2.0,label="вторая фирма")
plt.legend()
plt.grid()
plt.show()
```
